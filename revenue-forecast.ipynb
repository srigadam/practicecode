```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt1
```

```python
df_revenue= pd.read_csv('p21_bi_intern_test_revenues.csv')
df_revenue
```

```python
df_appointment= pd.read_csv('p21_bi_intern_test_appointments.csv')
df_appointment
```

```python
df=df_appointment.merge(df_revenue, on='appointment_id')
df
```

```python
df=df.sort_values('appointment_date')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
weights = [0.4, 0.3, 0.2]
total_weight = sum(weights)
normalized_weights = [w / total_weight for w in weights]
```

```python
df["WeightedMovingAvg"] = df['revenues'].rolling(window=len(normalized_weights)).apply(lambda x: (x * normalized_weights).sum())
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
X = df["appointment_date"].values.reshape(-1, 1)
y = df["WeightedMovingAvg"]
```

```python
df['appointment_date']=pd.to_datetime(df['appointment_date'])
```

```python
df.info()
```

```python
df['appointmet_month'] = df['appointment_date'].dt.month
```

```python
df
```

```python
df_1=df.groupby('appointmet_month',as_index=False)['revenues'].sum()
```

```python
df_1
```

```python
df_1.columns
```

```python
df_1["WeightedMovingAvg"] = df_1['revenues'].rolling(window=len(normalized_weights)).apply(lambda x: (x * normalized_weights).sum())
```

```python
print(df_1)
```

```python
df_1.dropna(inplace=True)
```

```python
df_1
```

```python
X=df_1['appointmet_month'].values.reshape(-1, 1)
y=df_1['WeightedMovingAvg']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
start_date = 4   # Starting date for the forecast
end_date = 12     # Ending date for the forecast
forecast_dates = range(start_date, end_date+1)
forecast_X = pd.DataFrame(forecast_dates, columns=["appointmet_month"])
forecast_y = model.predict(forecast_X)
```

```python
forecast_df = pd.DataFrame({"month": forecast_dates, "ForecastedRevenue": forecast_y})
print(forecast_df)
```
